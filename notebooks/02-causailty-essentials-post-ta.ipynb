#### Running the first regression

To run simple regressions we will youse the package *statsmodels*. We start by running a simple regression of median income on the percentage of adults with a college degree:

<center>$income_i = \beta_0 + \beta_1 college_i + \varepsilon_i$</center>

```python
import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col

res = smf.ols("income ~ college", data=df).fit()
res.summary()
```

**Causal Interpretation:** The main coefficient of interest is the one on $college$. This result implies that, on average, an increase by 1 percentage point in college graduates increases median household income by \$1'115

```python
sns.regplot(x, y)
```

One potential source of bias could be the number of adults with a high school degree *because having a high school degree may increase the likelihood of going to college and lead to higher income at the same time*. Luckly, we have the data on high school graduates and we can control for this in our regression.

```python
res_con1 = smf.ols("income ~ college + high", data = df).fit()
res_con1.summary()
```

**How did the coefficient on $college$ change? Did it change in the direction you expected?**

*The coefficient on $college$ is lower in magnitude when controlling for the share of adults with a high school degree. This hints that the coefficient from the previous regression is likely to be biased upward, as we would expect if $high$ were positively correlated with both the number of college graduates and income.*

**Why could it not be a good idea to include the percentage of high school graduate in the regression?**

- *High school completion in the U.S. is mandatory, so it is likely that the number of high school graduates does not affect the number of adults enrolling or completing college. Therefore, including the number of high school graduates in the regression might not account for any bias.*
- *The percentage of high school graduates may also act as a mediator. If we consider the conjecture that children with high educated parents are more likely to graduate from high school, and that high school degree leads to higher income, then the percentage of adults with a high school degree may act as a mediator.*


- expanse
- sat scores

**What other variables would you include?**

- *per pupil expenditure in education: the higher expenditure in education, the higher the likelihood that individuals are able of finishing college and the larger the number of college graduates. At the same time, higher education expenditure may help to foster better skills that result in higher income. However, the more college graduates in a region, the more likely it is that in this region the population values education and invests more in it, meaning that education expenditure may also be interpreted as a collider.*
- *region: this is the safest variable to include in the regression, because the effect it has on college and income is through pre-determined characteristics (like culture or historical background), that is features that are not affected by the current level of education and income.*

```python
res_con2 = smf.ols("income ~ college + C(region)", data=df).fit()
res_con2.summary()
```

```python
print(summary_col([res,res_con1, res_con2]))
```

**What are other problems that come up when using this regression to answer the initial research question?**

#### Estimation

```python
res = smf.ols("wage ~ C(collgrad)", data=df_nomiss).fit()
res.summary()
```

**Causal Interpretation:** On average, having a college degree increases hourly wages by \$3.6

```python
res1 = smf.ols("wage ~ C(collgrad) + C(smsa)", data=df_nomiss).fit()
res2 = smf.ols("wage ~ C(collgrad) + C(smsa) + C(south)", data=df_nomiss).fit()
res3 = smf.ols("wage ~ C(collgrad) + C(smsa) + C(south) + C(race)", data=df_nomiss).fit()
```

```python
print(summary_col([res,res1,res2,res3]))
```

### Residuals

So far we have assumed *homoskedastic* errors, meaning that we assumed the variance of the error term in our regression to be the same across observations and to be independent on the values of the explanatory variable. To investigate if this is the case, we need to calculate the residuals:

<center>$e_i = y_i - \hat{y}_i$ </center>

where $\hat{y}_i$ are the predicted values of the wage. Once we have the residuals, one way to investigate if the errors are homoskedastic or not is to plot the fitted value of the dependent variable against the residuals form the regression.

```python
yhat = res.predict()
resid = y - yhat

resid_plot = sns.scatterplot(resid, yhat)
resid_plot.set(xlabel='Residuals', ylabel='Predicted Wage')
plt.show()
```

```python
plt.hist(resid, bins=30, density=True)
plt.xlabel("Residuals", size=12)
plt.ylabel('Density', size=12)
plt.show()
```

The errors from our regression present evidence of mild heteroskedasticity. Therefore, to correctly estimate the causal effect of interest, we need to corrct for it by estimating the covariance matrix in a *robust* way. 

```python
res_rob = smf.ols("wage ~ C(collgrad)", data=df_nomiss).fit(cov_type='HC1')
print(summary_col([res,res_rob]))
```
